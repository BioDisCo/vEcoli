## Tetracycline Model Validation
### Transport
#### Background
Our model assumes that accumulation of tetracycline in the cytoplasm is driven by:
1. Difference in concentration between the interior and exterior of the cell
2. Electrostatic forces generated by the Donnan potential across the outer membrane

The flux is modeled in accordance with the [Nernst-Planck equation](https://en.wikipedia.org/wiki/Nernst%E2%80%93Planck_equation) with the following simplifying assumptions:
1. The periplasm and inner membrane are ignored since it is the outer membrane permeability gives rise to most of the variability in steady state accumulation ([Richter et al. 2017](https://doi.org/10.1038/nature22308)). 
2. Tetracycline reaches steady-state accumulation nearly instantaneously (in one timestep). Real experiments show that this process occurs over the span of a few minutes ([Thanassi et al. 1995](https://doi.org/10.1128/jb.177.4.998-1007.1995); [McMurry et al. 1982](https://doi.org/10.1128/AAC.22.5.791)), which is not drastically different on the scale of hour-long simulations.
3. The electrostatic force from the Donnan potential acts upon the concentration of tetracycline inside the membrane (more specifically the porins in the membrane), which is represented as the average of the internal and external concentrations.

In reality, many other factors contribute to the accumulation process, including pH differential across the inner membrane, chelation of tetracycline by magnesium, and binding of tetracycline to membranes and ribosomes ([Thanassi et al. 1995](https://doi.org/10.1128/jb.177.4.998-1007.1995); [Yamaguchi et al. 1991](https://doi.org/10.1128/AAC.35.1.53); [Argast and Beck 1985](https://doi.org/10.1007/BF00408069)).


#### Steady-State Accumulation
Given that susceptible E. coli cells appear to reach a steady-state internal tetracycline concentration about 15 times higher than external concentration ([McMurry et al. 1982](https://doi.org/10.1128/AAC.22.5.791)), the internal concentration of unbound tetracycline inside each cell was estimated to be about 8.4 times higher than the external concentration.

Our model falls about **3.5 times short** of this experimentally determined accumulation factor.

#### Knockout of Porin
Tetracycline is known to primarily enter E. coli cells through the outer membrane protein ompF. Strains with ompF knocked out have been previously shown to approximately halve tetracycline accumulation ([Richter et al. 2017](https://doi.org/10.1038/nature22308); [Thanassi et al. 1995](https://doi.org/10.1128/jb.177.4.998-1007.1995)) and double MIC ([Nikaido and Pag√©s 2012](https://doi.org/10.1111/j.1574-6976.2011.00290.x); [McMurry et al. 1982](https://doi.org/10.1128/AAC.22.5.791)).

#### Knockout of Active Efflux
Tetracycline is excreted from susceptible E. coli cells primarily via the acrAB-tolC efflux pump. Knocking out acrB (an integral and necessary component of this pump) resulted in only a slight increase tetracycline accumulation ([Feng et al. 2020](https://www.frontiersin.org/articles/10.3389/fmicb.2020.00556/full)) but a dramatic 75% reduction in MIC ([Viveiros et al. 2005](https://doi.org/10.1128/AAC.49.8.3578-3582.2005)).
